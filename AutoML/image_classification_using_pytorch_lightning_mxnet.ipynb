```
from torchvision.datasets import ImageFolder
import torch
from torch.utils.data import WeightedRandomSampler, DataLoader, random_split
import torchvision
from torchvision import datasets, models, transforms
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import pytorch_lightning as pl
import time
import copy
```

```
full_dataset = '/content/obuasi_labeled/neg'
def train_val_split(full_dataset, p_train = 0.7, generator = torch.Generator().manual_seed(42)):
  train_size = int(p_train * len(full_dataset))
  val_size = len(full_dataset) - train_size
  train_dataset, val_dataset = random_split(
      full_dataset, [train_size, val_size],
      generator=generator)
  return train_dataset, val_dataset


class MapDataset(torch.utils.data.Dataset):
    """
    Given a dataset, creates a dataset which applies a mapping function
    to its items (lazily, only when an item is called).

    Note that data is not cloned/copied from the initial dataset.
    """

    def __init__(self, dataset, map_fn):
        self.dataset = dataset
        self.map = map_fn

    def __getitem__(self, index):
        if self.map:
            x = self.map(self.dataset[index][0])
        else:
            x = self.dataset[index][0]  # image
        y = self.dataset[index][1]   # label
        return x, y

    def __len__(self):
        return len(self.dataset)


class ChipsDataModule(pl.LightningDataModule):

      def __init__(self, root_folder='/content/obuasi_gsh_chips/obuasi_untagged', p_train=0.7, batch_size=32, random_seed=1):
          super().__init__()
          self.root_folder = root_folder
          self.batch_size = batch_size
          self.p_train = p_train
          self.generator = torch.Generator().manual_seed(random_seed)

      # When doing distributed training, Datamodules have two optional arguments for
      # granular control over download/prepare/splitting data:

    
      def get_transforms(self):
        train_transform = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(),
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize(
              mean=[0.485, 0.456, 0.406],
              std=[0.229, 0.224, 0.225],
              inplace=True)
        ])

        val_transform = transforms.Compose([
            transforms.Resize(224),
            transforms.ToTensor(),
            transforms.Normalize(
              mean=[0.485, 0.456, 0.406],
              std=[0.229, 0.224, 0.225],
              inplace=True)
        ])

        return train_transform, val_transform

      # OPTIONAL, called for every GPU/machine (assigning state is OK)
      def setup(self, stage):
          chips_dataset = ImageFolder(root=self.root_folder)
          train_dataset, val_dataset =train_val_split(
              chips_dataset, p_train=self.p_train, generator=self.generator)

          train_transform, val_transform = self.get_transforms()

          self.train_dataset = MapDataset(train_dataset, train_transform)
          self.val_dataset = MapDataset(val_dataset, val_transform)

          # instance weigths for dataloaders
          labels = pd.Series([i[1] for i in self.train_dataset])
          counts = labels.value_counts()

          self.class_weights = (sum(counts)/counts)
          self.weights = labels.apply(lambda l: self.class_weights[l])

      # return the dataloader for each split
      def train_dataloader(self):
          sampler = WeightedRandomSampler(
              weights=self.weights, num_samples=len(self.weights), generator=self.generator)
          train_dataloader = DataLoader(
              self.train_dataset, batch_size=self.batch_size, sampler=sampler)
          return train_dataloader

      def val_dataloader(self):
          val_dataloader = DataLoader(self.val_dataset, batch_size=self.batch_size)
          return val_dataloader
```

```
class LitModel(pl.LightningModule):

    def set_parameter_requires_grad(self):
        if self.feature_extract:
            for param in self.model.parameters():
                param.requires_grad = False

    def __init__(
        self, num_classes, 
        class_weights=None, use_pretrained=True, feature_extract = True, lr=1e-5):
        super().__init__()
        self.feature_extract = feature_extract
        self.criterion = nn.CrossEntropyLoss(weight=class_weights)

        self.model = models.resnet34(pretrained=use_pretrained)
        self.lr=lr

        self.set_parameter_requires_grad()

        num_ftrs = self.model.fc.in_features
        self.model.fc = nn.Linear(num_ftrs, num_classes)

        self.input_size = 224

    def forward(self, x):
        x = self.model(x)
        return x

    def configure_optimizers(self):
        print("Params to learn:")
        if self.feature_extract:
            params_to_update = []
            for name, param in self.model.named_parameters():
                if param.requires_grad == True:
                    params_to_update.append(param)
                    print("\t",name)
        else:
            params_to_update = self.model.parameters()
            for name,param in model.named_parameters():
                if param.requires_grad == True:
                    print("\t",name)

        optimizer = optim.Adam(params_to_update, lr=self.lr, betas=(0.9, 0.999), eps=1e-08, weight_decay=1e-5)
        return optimizer

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = self.criterion(y_hat, y)

        roc_auc_fn = pl.metrics.classification.AUROC()

        y_pred = F.softmax(y_hat, dim=1)[:, 1]
        roc_auc = roc_auc_fn(y_pred, y)

        result = pl.TrainResult(loss)
        result.log('train/loss', loss, logger=True, on_epoch=True)
        result.log('train/roc_auc', roc_auc, logger=True, on_epoch=True)

        return result

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = self.criterion(y_hat, y)
        result = pl.EvalResult()
        result.log('val/loss', loss, logger=True)
        # print('eval_step: {}'.format(result))
        y_pred = F.softmax(y_hat, dim=1)[:, 1]

        result.y = y
        result.y_pred = y_pred

        return result

    def validation_epoch_end(self, outputs):
        y = outputs.y
        y_pred = outputs.y_pred
        val_loss = outputs['val/loss'].mean()

        f1_fn = pl.metrics.classification.F1()
        roc_auc_fn = pl.metrics.classification.AUROC()
        cm_fn = pl.metrics.classification.ConfusionMatrix()

        threshold = 0.5
        
        f1_score = f1_fn(y_pred > threshold, y)
        roc_auc = roc_auc_fn(y_pred, y)
        cm = cm_fn(y_pred > threshold, y)

        # print('eval_end :{}'.format(result))
        # self.logger.experiment.add_text('val/cm', str(cm), global_step=self.global_step)
        self.logger.experiment.add_pr_curve('val/pr_curve', y, y_pred, global_step=self.global_step)

        result = pl.EvalResult(val_loss, checkpoint_on=-roc_auc)
        result.log('val/roc_auc', roc_auc)
        result.log('val/loss', val_loss)
        result.log('val/f1', f1_score)

        return result
```
