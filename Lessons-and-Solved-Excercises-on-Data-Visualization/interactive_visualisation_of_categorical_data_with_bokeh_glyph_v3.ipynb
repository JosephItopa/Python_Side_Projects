```python
from bokeh.sampledata.autompg import autompg_clean as cars
```

```python
cars.head()
```

```python
cars_by_origin = cars[['origin','name']].groupby('origin').count()
cars_by_origin.rename(columns = {'name':'count'}, inplace = True)
cars_by_origin.reset_index(inplace=True)
```

```python
cars_by_origin
```

```python
from bokeh.models import ColumnDataSource
cars_by_origin_src = ColumnDataSource(cars_by_origin)
```

```python
factors = cars['origin'].drop_duplicates().to_list()
```

```python
factors
```

```python
plot = figure(
      x_range = factors, title = "cars by origin",
      plot_width = 800, plot_height = 300,
      toolbar_location = None, tools =""
)
plot.vbar(x='origin', top = 'count', source = cars_by_origin_src, width = 0.9)
plot.title.text_font_size = "12pt"
plot.y_range.start = 0
plot.xgrid.grid_line_color = None
show(plot)
```

```python
from bokeh.transform import factor_cmap
from bokeh.palettes import Colorblind3
```

```python
cat_color_map = factor_cmap(field_name = 'origin', palette = Colorblind3, factors = factors)
```

```python
plot = figure(
      x_range = factors, title = "cars by origin",
      plot_width = 800, plot_height = 300,
      toolbar_location = None, tools =""
)
plot.vbar(x='origin', top = 'count', source = cars_by_origin_src, width = 0.9, legend_field = "origin", line_color = 'white', fill_color = cat_color_map)
plot.title.text_font_size = "12pt"
plot.y_range.start = 0
plot.xgrid.grid_line_color = None
plot.legend.orientation = 'horizontal'
plot.legend.location = 'top_center'
plot.y_range.end = 400
show(plot)
```

```python
cars_by_origin_cyl = cars[['origin', 'name', 'cyl']].groupby(['origin', 'cyl']).count()
cars_by_origin_cyl.rename(columns={'name':'count'}, inplace = True)
cars_by_origin_cyl.reset_index(inplace = True)

cars_by_origin_cyl['cyl'] = cars_by_origin_cyl['cyl'].astype('str')

pivoted_table = cars_by_origin_cyl.pivot_table(
    index = 'origin',
    columns = 'cyl',
    values = 'count',
    aggfunc = 'sum',
    fill_value = 0
).rename_axis(None, axis = 1).reset_index()
```

```python
pivoted_table
```

```python
data = pivoted_table.to_dict('list')
origin_list = data['origin']
cyl_list = list(pivoted_table)[1:]
```

```python
from bokeh.palettes import Category20_5
plot = figure(
    x_range = origin_list,
    plot_height = 300,
    plot_width = 450,
    title = "Cars by origin and Cyl",
    toolbar_location = None, tools = ""
)
plot.vbar_stack(
    stackers = cyl_list,
    x = 'origin',
    color = Category20_5,
    legend_label = cyl_list,
    width = 0.9
)

plot.y_range.start = 0
plot.x_range.range_padding = 0.1
plot.xgrid.grid_line_color = None
plot.axis.minor_tick_line_color = None
plot.outline_line_color = None
plot.legend.location = "top_left"
plot.legend.orientation = "horizontal"
show(plot)
```
