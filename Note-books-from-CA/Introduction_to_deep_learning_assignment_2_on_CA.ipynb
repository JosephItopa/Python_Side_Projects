```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

Data source: https://www.kaggle.com/mustafaali96/weight-height

```python
df_wh = pd.read_csv('/content/weight-height.csv')
```

```python
df_wh.sample(5)
```

```python
df_wh['Gender'].unique()
```

```python
df_wh.plot(
        kind='scatter',
        x = 'Weight',
        y = 'Height',
        title='Plot of Height against Weight',
        figsize = (8, 6)
        )
```

```python
#filtered_df = df.loc[(df['date'] >= '2020-09-01') & (df['date'] < '2020-09-15')]
df_males = df_wh[df_wh['Gender'] == 'Male']
df_females = df_wh.query('Gender == "Female"')
fig, ax = plt.subplots()

df_males.plot(kind='scatter',
              x = 'Height',
              y = 'Weight',
              ax = ax, color = 'blue', alpha = 0.3,
              title = 'Male & Female Populations'
              )
df_females.plot(kind='scatter',
              x = 'Height',
              y = 'Weight',
              ax = ax, color = 'red', alpha = 0.3
              )
```
