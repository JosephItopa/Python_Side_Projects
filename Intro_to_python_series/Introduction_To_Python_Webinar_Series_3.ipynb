##  **IV)  Generators and Decorators**
    

Whenever a call is made to a python function, it make use of the current argument passed to it and then, after answering the call, it wipes its memory in another to accept another argument from a call. Using the concept of generators, we can make the function remember the arguments.

Instead of using **return** keyword in the function, we use **yield** which aids it to remember.

```
def infinite_seq():
    num = 0
    while True:
        yield num
        num += 2


for i in infinite_seq():
     print(i, end=" ")
```

Decorator argument has the ability to alter the behaviour of other functions or method.

```
# declare decorator
def smart_sum(funny):
    def inner_val(a, b):
        print("I am going to sum", a, "and", b)
        if b == 0:
            print("Lets sum")
            return
        return funny(a, b)
    return inner_val

@smart_sum # call decorator
def sum(a, b):
    print(a + b)

sum(2,5)    
```
