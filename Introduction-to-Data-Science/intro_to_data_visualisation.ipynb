```python
# Load libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv('/content/winequality_red.csv')
```

```python
# randomly display 5 rows 
df.sample(5)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# list columns
df.columns
```

**Line and dot(or scatter) plots**

```python
# A simple line plot using seaborn 
sns.lineplot(data=df, x="alcohol", y="quality")
```

```python
# simple line plot with grid lines using seaborn
sns.set_theme(style="darkgrid")
sns.lineplot(
    data=df, x="citric acid", y="quality"
    )
```

```python
# load weight-height dataset
df_wh = pd.read_csv('/content/weight-height.csv')
```

```python
# randomly display 5 rows
df_wh.sample(5)
```

```python
# check for missing values
df_wh.isnull().sum()
```

```python
# drop missing values
df_wh = df_wh.dropna()
# convert float data types to integer
df_wh.Weight = df_wh['Weight'].astype('int64')
df_wh.Height = df_wh['Height'].astype('int64')
```

```python
# a scatterplot using seaborn
sns.relplot(x="Height", y="Weight", hue="Gender", data=df_wh)
```

**Box plot**

```python
# statistical details of the dataset
df.describe()
```

```python
# a simple box plot using seaborn
sns.boxplot( y=df["quality"])
plt.show()
```

```python
# using seaborn for many boxplot on a chart
sns.boxplot(x="Height", y="Weight", hue="Gender", data=df_wh)
```

**Histogram and pair plots**

```python
# using matplotlib to plot histogram
plt.hist(x=df['alcohol'])
plt.title('Histogram distribution for alcohol content')
plt.show()
```

```python
# using matplotlib to plot histogram of just two features
plt.hist(x=df_wh['Gender'])
plt.title('Histogram distribution for Gender')
plt.show()
```

```python
# using seaborn and matplotlib to plot a histogram of two features
sns.histplot( x=df_wh["Gender"], y=df_wh["Weight"])
plt.xlabel('Gender')
plt.ylabel('Weight')
plt.title('Histogram distribution for Weight')
plt.legend()
plt.show()
```

```python
# using the pairplot api from seaborn to show the relationship between the features of the dataset
sns.pairplot(df_wh)
```

```python
# Histogram plot with 30 bins using seaborn
sns.histplot(data=df_wh, x="Weight", bins=30)
```

```python
# Histogram plot with kernel density using seaborn
sns.histplot(data=df_wh, x="Height", bins=30, kde=True)
```
